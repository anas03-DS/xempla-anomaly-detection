```python
X_scaled = scaler.fit_transform(X) #Learns the average and standard deviation of each sensor

#Converts all features so they have a mean of 0 and std of 1
```

```python
from sklearn.ensemble import IsolationForest
```

```python
# applying Isolation Forest
model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
```

```python
df['anomaly'] = model.fit_predict(X_scaled)

#expected output = -1 for anamoly
# 1 for normal
```

```python
# try to covert outputu to 1 and 0
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})
 # 0 means ok
 # 1 means problem detected
```
